```python
import pandas as pd
import random
```

```python
# Replace 'your_input_file.csv' with the path to your CSV file
input_file = '/content/vicuna_outputs_ritika.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(input_file)
```

```python
df.head()
```

```python
# extract questions
qindices = []
# extract statements
sindices = []
for i,x in enumerate(df.instruction.values):
    if '?' in x:
        qindices.append(i)
    else:
        sindices.append(i)
```

```python
len(sindices)
```

```python
len(qindices)
```

```python
# df with only questions
dfq = df.iloc[qindices,:]
dfq.reset_index(drop=True, inplace=True)
```

```python
# df with only statements
dfs = df.iloc[sindices,:]
dfs.reset_index(drop=True, inplace=True)
```

```python
# select 175 random questions
dfq = dfq.sample(n = 175)
```

```python
dfq.reset_index(drop=True, inplace=True)
```

```python
# select 175 random statements
dfs = dfs.sample(n = 175)
dfs.reset_index(drop=True, inplace=True)
```

```python
df_merged = dfq.append(dfs, ignore_index=True)
df_merged
```

Generate 3 datasets - without reason, with trunc reasons, with full reasons

```python
## without reasons

# Specify the file path where you want to save the CSV
csv_file_path = './harmfulprompts350noreason.csv'

noreason = df_merged[['instruction','input','output']]

# Save the DataFrame as a CSV file
noreason.to_csv(csv_file_path, index=False, header=True)

print(f'CSV file saved as {csv_file_path}')
```

```python
## with long reasons

# Specify the file path where you want to save the CSV
csv_file_path = './harmfulprompts350allreason.csv'

fullreason = df_merged[['instruction','input','out']]

# Save the DataFrame as a CSV file
fullreason.to_csv(csv_file_path, index=False, header=True)

print(f'CSV file saved as {csv_file_path}')
```

Some processing for short reasons

```python
# to find the min number of sentences
min = 100
for i,x in enumerate(df_merged.out.values):
    s = x.split('.')
    if min>len(s):
        min = len(s)

min
```

```python
def truncate(x):
    s = x.split('.')
    if len(s)==3:
        return s[0]
    else:
        sup = '.'.join(s[:2])
        return sup + '.'

df_merged['out2'] = df_merged['out'].apply(truncate)
```

```python
## with short reasons

# Specify the file path where you want to save the CSV
csv_file_path = './harmfulprompts350somereason.csv'

somereason = df_merged[['instruction','input','out2']]

# Save the DataFrame as a CSV file
somereason.to_csv(csv_file_path, index=False, header=True)

print(f'CSV file saved as {csv_file_path}')
```
