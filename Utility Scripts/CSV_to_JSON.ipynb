```python
import csv
import json
```

### Read the CSV File:
Use the csv.reader to read the data from the CSV file. You can open the CSV file using open and then read the rows.

```python
csv_file = '/content/harmfulprompts350noreason.csv'

csv_data = []
with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        csv_data.append(row)
```

```python
csv_data
```

### Create a List of Dictionaries:
In order to convert the CSV data to JSON, create a list of dictionaries. Each row in the CSV file will become a dictionary, where the keys are the column headers (from the CSV file's first row) and the values are the corresponding values from each row.

```python
# Extract column headers from the first row of the CSV file
headers = ['instruction','input','output']
print(headers)
# Initialize an empty list for JSON data
json_data = []

print(csv_data[1])
# Iterate through the rows (excluding the first row with headers)
for row in csv_data[1:]:
    # Create a dictionary for each row
    row_dict = {}
    for i in range(len(headers)):
        row_dict[headers[i]] = row[i]
    json_data.append(row_dict)
```

### Write the JSON Data to a File:
Finally, write the JSON data to a JSON file using the json.dump method.

```python
json_file = 'alpaca_data.json'

with open(json_file, 'w') as file:
    json.dump(json_data, file, indent=4)
```

## Shuffle a JSON file

```python
import random
import json
import ast

# Load the JSON data from the file
input_file = '/content/alpaca_data.json'
output_file = 'alpaca_data_harmful_shuffled.json'

jsondata = ast.literal_eval(json.dumps(input_file))

with open(input_file, 'r') as file:
    data = json.load(file)

# Shuffle the data
random.shuffle(data)

# Write the shuffled data back to the JSON file
with open(output_file, 'w') as file:
    json.dump(data, file, indent=4)  # Use indent for pretty formatting, if desired
```

## Shuffle a CSV file

```python
import csv
import random

# Specify the input and output file paths
input_file_path = '/content/output_vicuna_wizard.csv'
output_file_path = '/content/output_vicuna_wizard_shuffled.csv'

# Read the CSV file into a list of dictionaries
with open(input_file_path, 'r') as csv_file:
    reader = csv.DictReader(csv_file)
    data = list(reader)

# Shuffle the list of dictionaries
random.shuffle(data)

# Write the shuffled data to a new CSV file
with open(output_file_path, 'w', newline='') as csv_output:
    fieldnames = reader.fieldnames
    writer = csv.DictWriter(csv_output, fieldnames=fieldnames)

    # Write header
    writer.writeheader()

    # Write shuffled data
    for row in data:
        writer.writerow(row)

print(f"Records shuffled and saved to '{output_file_path}'.")
```
